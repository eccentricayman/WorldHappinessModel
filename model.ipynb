```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Loading the dataset
data = pd.read_csv("world-happiness-report-2021.csv")
#Add extra datasets if needed for multiple etc....
data.dropna(axis=1)
```

```python
#First Linear Regression Vectors using IMDB
y = np.array(data["Social support"])
x = np.array(data["Ladder score"])
```

```python
#fit_Linear 
def fit_linear(x,y):
    xm = np.mean(x)
    ym = np.mean(y)
    syy = np.mean((y-ym)**2)
    syx = np.mean((y-ym)*(x-xm))
    sxx = np.mean((x-xm)**2)
    b1 = syx/sxx
    b0 = ym - b1*xm
    yhat=b0+b1*x
    rss = np.sum((y-yhat)**2)
    tss = np.sum((y-ym)**2)
    rsq = 1-rss/tss
    return b0, b1, rsq
```

```python
b0, b1, rsq = fit_linear(x,y)
yHat = b0+b1*x
```

```python
plt.scatter(x,y)
plt.plot(x, yHat)
plt.xlabel("Ladder Score")
plt.ylabel("Social Support")
plt.show()
```
